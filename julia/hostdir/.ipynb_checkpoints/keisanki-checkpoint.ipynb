## 問1
#### (1)

````julia
import LinearAlgebra, IterativeSolvers, Plots, PyPlot
using LinearAlgebra, Plots

function newton(x0,y0)
    f_1(x,y)=y^2-x^3+4x
    f_2(x,y)=(x/2)^4+y^4-1
    Df_1_x(x,y)=-3x^2+4
    Df_1_y(x,y)=2y
    Df_2_x(x,y)=(x/2)^3
    Df_2_y(x,y)=4y^3
    f(x,y)=[f_1(x,y); f_2(x,y)]
    Df(x,y)=[Df_1_x(x,y) Df_1_y(x,y); Df_2_x(x,y) Df_2_y(x,y)]
    x=x0
    y=y0
    TorF=0
    
    for i in 1:100000
        if det(Df(x,y))==0
            print("正則でない.ループ回数: ",i)
            break
        end
        (z, w)=[x; y]-Df(x,y)\f(x,y)
        if norm([x; y]-[z; w], 2)<10^(-10)
            print("収束した.ループ回数: ",i)
            TorF=1
            break
        end
        x=z
        y=w
    end
    return [x; y; TorF]
end
````

##### newton法の実装の説明
与えられた2式とそれぞれをx,yで微分した式をf_i,Df_i_x,Df_i_yのように定義し、それらをつかってf,Dfという行列を定義した。
初期値をx0,y0として与え、(z, w)=[x; y]-Df(x,y)\f(x,y)をfor文で何度も実行することでnewton法を実装している。<br>
ただし、ループ内で(z, w)への代入が行われる前にDf(x,y)の行列式をLinerAlgebraを用いて判定し、0であった場合はループを抜ける。<br>
また、代入前後の2変数ベクトルのL2距離を用いて計算し、10^(-10)未満となったときも収束が完了したとしてループを抜ける。<br>
for文は100000回ループとしているが、これは初期値が遠い解と遠い場合収束しづらいことを利用し、ループが最後まで完了することで少なくとも初期値と解が近くはないことを示している。<br>
また、TorFという値を返しているが、これは100000回以内に収束した場合Tを返す。

#### (3)
まずfor文を用いて、-5から5の範囲のx,yに対して繰り返しnewton法を適用し、収束先をX,Y、収束した点をX_T,Y_T、収束しなかった点X_F,Y_Fとして収集する。

````julia
X=[]
Y=[]
X_F=[]
Y_F=[]
X_T=[]
Y_T=[]
for i in 0:20
  for j in 0:20
    (x,y,tf)=newton(0.5i-5, 0.5j-5)
    push!(X,x)
    push!(Y,y)
    if tf == 0
            push!(X_F,0.5i-5)
            push!(Y_F,0.5j-5)
        else
            push!(X_T,0.5i-5)
            push!(Y_T,0.5j-5)
    end
  end
end
````

まず収束先をplotする

````julia
scatter(X, Y, xlabel="X", ylabel="Y", legend=false)
````

次に収束した点をplotする

````julia
scatter(X_T, Y_T, xlabel="X_T", ylabel="Y_T", legend=false)
````

````julia
scatter(X_F,Y_F,xlabel="X_F",ylabel="Y_F",legend=false)
````

これらの結果から、y=0からプログラムを始めた場合や、収束先から離れた点では収束せず、収束先はy=0,x=2付近に集中している。また、(±2,±0.5),(0,±1.0)付近とy=0,x=整数の点も収束先である可能性が高い。しかし、実際にはy=0のときの厳密解としてx=±2であり、収束先と厳密解が少なくともy=0の場合一致していない。

まず、題の3重対角行列を生成する関数を定義する。

````julia
import LinearAlgebra, IterativeSolvers, Plots, PyPlot
using LinearAlgebra
function generate(n,c)
    A=Matrix{Int}(I, n, n)
    for i in 1:n
        for j in 1:n
            if i==j
                A[i,j]=A[i,j]+1
              elseif abs(i-j)==1
                A[i,j]=A[i,j]-1
            end
        end
    end
    A[1,1]=c
    return A
end
````

n=5,c=2の行列Aを生成し、wigvalsを用いて固有値を出力する。

````julia
A=generate(5,2)
````

````julia
eigvals(A)
````

#### (2)

eigvalsの出力結果が正しいとするとA_5は正定値であることがわかる。また、定義より対称であるのでCG法が使える。

````julia
function cg(x0,n,c,b)
    A=generate(n,c)
    Ac=eigvals(A)
    for k in 1:n
        if 0>=Ac[k]
            return("∅","∅")
        end
    end
    r=b-A*x0
    p=r
    x=x0
    for i in 1:1000
        a=(p'*r )/((A*p)'*p)
        if norm(x-x+a*p,2)<10^(-10)
            return (x,i)
        end
        x=x+a*p
        r=r-a*A*p
        b=((A*p)'*r)/((A*p)'*p)
        p=r-b*p
    end
end
cg([0;0;0;0;0],5,2,[0;0;0;0;6])
````

ここで、return(x,i)のうちxが解でiが繰り替えした回数である。<br>
結果より[1;2;3;4;5]が解だと考えられる。
実際、Ax=bは成り立つ。

## 問3
#### (1)

Nを与えたときに複合台形公式を計算する関数Trapezoidalを以下のように定義する。

````julia
function Trapezoidal(N)
    h=(pi-0)/N
    x=0
    for n in 1:N
        x=x+(h/2)*((sin(h*(n-1)))+sin(h*n))
    end
    return x
end
````

区間[a,b]で2点型Gauss数値積分公式を計算する関数Gaussを以下のように定義する。次いで、N分割の複合公式を計算する関数composeを定義する

````julia
function Gauss(a,b)
    x0=-1/sqrt(3)
    X1=1/sqrt(3)
    w0=1
    w1=1
    A=((b-a)/2)*(w0*sin(((b-a)*x0)/2+(a+b)/2)+(w1*sin(((b-a)*x0)/2+(a+b)/2)))
    return A
end

function compose(N)
    G=0
    i=0
    for n in 1:N
        j=i+(pi/N)
        G=G+Gauss(i,j)
        i=j
    end
    return G
end
````

for文を使ってN=10,20,40,80のときの結果をTrapezoidalの値-Gaussの値=誤差の差、という風に表示していく。

````julia
for N in 0:3
    T=Trapezoidal(10*2^N)
    G=compose(10*2^N)
    println(T,"-",G,"=",T-G)
end
````

## 問4
#### (1)
変数分離を使って解く。<br>
du/u+1=dtの両辺を積分し<br>
log|u+1|=t+C(積分定数）とする。u(0)=1よりC=log2である。<br>
よってu(t)=exp(t+log2)-1
#### (2)
時間刻み幅を1/hとし、求める点たときhを引数とした関数Rungeを以下のように定義する。
表示される値は(数値的解-厳密解=誤差)である。
Runge(20),Runge(40),Runge(80)では結果が長すぎたため省略したが、誤差は0.6981347736120096,0.3589765654023118,0.18207042762718118であった。

````julia
function Runge(h)
    u=1 # u(0)=1
    println("u0=",u)
    for i in 1:2h
        u=u+((1/h)*(u+1))
        uu=exp(((1/h)*i)+log(2))-1
        println("u",i," : ",u,"-",uu,"=",u-uu)
    end
end
Runge(10)
````

````julia
Runge(20)
````

````julia
Runge(40)
````

````julia
Runge(80)
````

#### (3)
初期値が定まっていたとして、hが小さいほどu(x0+h)は初期値に近い値をとり、前進Euler法で使っている式の誤差も少なくなる。<br>
特に、一次式を使って近似しているので、基準の点からの距離に単純に比例して誤差が増えるはずである。<br>
runge-kuttaでは誤差の含まれた値を新たな基準にしているため誤差も線形に比例すると考えられる。<br>
実験結果はそれらに従って、刻みを半分にすれば誤差も半分になっている。
